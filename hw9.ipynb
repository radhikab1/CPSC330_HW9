<hr>

rubric={points}

The following questions pertain to the lecture on survival analysis. We'll consider the use case of customer churn analysis.

1. What is the problem with simply labeling customers are "churned" or "not churned" and using standard supervised learning techniques?
2. Consider customer A who just joined last week vs. customer B who has been with the service for a year. Who do you expect will leave the service first: probably customer A, probably customer B, or we don't have enough information to answer? Briefly explain your answer. 
3. If a customer's survival function is almost flat during a certain period, how do we interpret that?

<div class="alert alert-warning">

Solution_1
    
</div>

_Points:_ 6

1. The problem with labeling customers as ‚Äúchurned‚Äù or ‚Äúnot churned‚Äù is that this ignores time-to-event and fails to handle right-censoring. Many customers have not yet churned at the time we observe them, so their true churn time is censored, not ‚Äúnon-churn.‚Äù Treating these censored observations as if the event will never occur introduces bias, and removing them leaves you with only customers whose churn events have already happened (those with shorter observed event times). This distorts the distribution of churn times and leads standard supervised learning models to produce incorrect or misleading predictions.

2. We cannot determine who is more likely to churn first, because survival analysis requires more than just the current observed time. The hazard of churn can increase, decrease, or stay constant over time depending on the service, and we have no information about the underlying hazard function or customer covariates. Therefore, based solely on ‚Äúone week vs. one year,‚Äù we do not have enough information to compare their time-to-event distributions or their hazards of churn.

3. If the survival function is nearly flat during some time interval, it means that the probability of surviving beyond that time, i.e., not churning barely changes. In other words, very few churn events occur in that period, and the hazard of churn is low. Customers who reach that interval tend to continue remaining with the service, indicating a stable retention phase.

# Predicting Credit Card Default Probabilities


Machine learning can improve banking at a low cost. ML models can quickly find patterns and make predictions, resulting in increased efficacy for banks and better security for clients. Our project was to predict whether a client will default on their credit card payment. This is a binary **classification** problem‚Äìdiscerning between defaulting or not. We tried different models, each with its own strengths and weaknesses, then selected the best-performing model.

### Wrangling and EDA
In order to use the data effectively and appropriately, we need to understand it well. Fortunately, the training data was already clean and tidy. It consists of 30,000 rows and 25 columns. Each row represents one client and includes: 

Our **target** variable, the **class**:
- default.payment.next.month
And **features**, the explanatory variables:
- Customer Id: 'ID'
- Credit Limit: 'LIMIT_BAL'
- Demographics:'SEX', 'EDUCATION','MARRIAGE','AGE'
- Recent repayment statuses: 'PAY_0','PAY_2',...,'PAY_6'
- Recent bill amounts: 'BILL_AMT1','BILL_AMT2',...,'BILL_AMT6'
- Recent payment amounts: 'PAY_AMT1', 'PAY_AMT2',...,'PAY_AMT6'


We conducted EDA to examine the data. We looked at summary statistics, data types, and some distributions.

We want to know if there is **class imbalance**, to know if we would need to take extra precautions. Thankfully, this chart shows us the imbalance is not severe. In this context, a false negative would be more harmful than a false positive, but we still want to avoid false positives.

This helped us select **ROC-AUC** score as our evaluation metric. Given that we wanted to compare different models against each other, ROC-AUC was the most suitable metric, because it provides a good summary of performance over all thresholds, while handling class imbalances.


### Feature Engineering and Preprocessing

Hoping that they would either bring new insights or reduce computational burden, we engineered new features: 
- average_bill_amt (average of 6 bill amounts)
- average_payment_amt (average of 6 payment amounts)
- avg_bill_to_pay_ratio (ratio between the two previous features)


We then preprocessed the data, applying transformations so each model receives consistent inputs.

We applied standard scaling to the numerical features to ensure equal consideration during training. We applied One Hot Encoding to the categorical features (sex, education, and marriage) to preserve their distinct meanings.

### Caveats and Limitations

#### Limited Tuning
Because this was a school project done with time and resource constraints, we were not able to run a very robust search for hyperparameters. While our best-performing model has decent results, there might be better-performing hyperparameter values. We might not be using the best hyperparameters, or even the best type of model. 

#### Demographic Biases
The features include demographic information‚Äìsex, marriage, and education. This might cause biases to show up in our model, which is particularly harmful for a context such as banking. For example, the model might learn that young people are at risk for default, and then penalize people just for being young. 

#### Uniform Data
The data comes from Taiwanese clients between April and September 2005. There might have been changes in patterns in the past 20 years, and the behaviour of Taiwanese clients might not generalize to other regions. This means that our model might perform poorly in other temporal and regional contexts.
